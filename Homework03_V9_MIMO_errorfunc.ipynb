## MIMO LSTM model on Spring data (Predicting Total Building and PV using previous Predicting Total Building, PV, Dry Bulb Temp, and Photometric)

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# -------------------------------
# Step 1: Prepare the Data
# -------------------------------

# For demonstration purposes, we generate dummy data.
# Replace these with your actual data:
n_samples = 83     # number of samples
n_timesteps = 7    # number of time steps (e.g., 7 days)
n_features = 96    # features per time step
n_targets = 48     # 48 predicted target variables per sample

# Dummy input: X_multivariate with shape (83, 7, 96)
X_multivariate = np.random.rand(n_samples, n_timesteps, n_features)

# Dummy target: combined_target_scaled with shape (83, 48)
combined_target_scaled = np.random.rand(n_samples, n_targets)

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_multivariate, combined_target_scaled, test_size=0.2, random_state=42
)

# -------------------------------
# Step 2: Define Custom Loss Functions
# -------------------------------

# 1. Root Mean Square Error (RMSE) Loss
def rmse_loss(y_true, y_pred):
    return tf.sqrt(tf.reduce_mean(tf.square(y_pred - y_true)))

# 2. R-squared Loss: We return 1 - R2 so that minimizing this loss pushes R² closer to 1.
def r2_loss(y_true, y_pred):
    # Sum of squares of residuals.
    ss_res = tf.reduce_sum(tf.square(y_true - y_pred))
    # Total sum of squares.
    ss_tot = tf.reduce_sum(tf.square(y_true - tf.reduce_mean(y_true)))
    # Compute R²; add a small epsilon to avoid division by zero.
    r2 = 1 - ss_res / (ss_tot + tf.keras.backend.epsilon())
    return 1 - r2  # minimizing 1 - R² is equivalent to maximizing R²

# 3. Coefficient of Variation of RMSE (CVRMSE) Loss
def cvrmse_loss(y_true, y_pred):
    rmse = tf.sqrt(tf.reduce_mean(tf.square(y_pred - y_true)))
    # Divide by the mean of y_true; add an epsilon to avoid division by zero.
    mean_true = tf.reduce_mean(y_true)
    return rmse / (mean_true + tf.keras.backend.epsilon())

# Note: For Mean Absolute Error (MAE) and Mean Squared Error (MSE),
# we can use the built-in loss functions by name ('mae' and 'mse').

# -------------------------------
# Step 3: Build the LSTM Model Function
# -------------------------------
def create_model():
    """Creates and returns a new instance of the LSTM model."""
    model = Sequential()
    model.add(LSTM(64, input_shape=(n_timesteps, n_features)))
    model.add(Dense(n_targets))
    return model

# -------------------------------
# Step 4: Train Models with Different Loss Functions
# -------------------------------

# Define a dictionary with the loss functions to try.
# For built-ins, we just use the string names.
loss_functions = {
    'mse': 'mse',          # baseline: Mean Squared Error
    'rmse': rmse_loss,     # custom Root Mean Square Error
    'mae': 'mae',          # Mean Absolute Error
    'r2': r2_loss,         # custom R-squared loss (1 - R²)
    'cvrmse': cvrmse_loss  # custom Coefficient of Variation RMSE
}

# Set up early stopping to prevent overfitting.
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Prepare dictionaries to store training history and trained models.
histories = {}
models = {}

# Train a new model for each loss function.
for loss_name, loss_func in loss_functions.items():
    print(f"\nTraining with {loss_name} loss...")
    # Create a new instance of the model
    model = create_model()
    # Compile the model with the given loss function and Adam optimizer.
    model.compile(loss=loss_func, optimizer='adam')
    # Train the model.
    history = model.fit(
        X_train, y_train,
        epochs=100,             # Adjust epochs as needed
        batch_size=16,          # Adjust batch size as needed
        validation_data=(X_val, y_val),
        callbacks=[early_stop],
        verbose=0  # set to 1 to see training progress
    )
    # Store the history and model for comparison.
    histories[loss_name] = history.history
    models[loss_name] = model
    print(f"Final validation loss for {loss_name}: {history.history['val_loss'][-1]:.4f}")

# -------------------------------
# Step 5: Evaluate Models on the Validation Set
# -------------------------------
evaluation = {}

for loss_name, model in models.items():
    # Predict on validation data
    y_pred = model.predict(X_val)
    mse_val = mean_squared_error(y_val, y_pred)
    rmse_val = np.sqrt(mse_val)
    mae_val = mean_absolute_error(y_val, y_pred)
    r2_val = r2_score(y_val, y_pred)
    cvrmse_val = rmse_val / np.mean(y_val)
    
    evaluation[loss_name] = {
        'MSE': mse_val,
        'RMSE': rmse_val,
        'MAE': mae_val,
        'R2': r2_val,
        'CVRMSE': cvrmse_val
    }

# Print the evaluation results.
print("\nEvaluation Metrics on Validation Set:")
for loss_name, metrics in evaluation.items():
    print(f"\nLoss Function: {loss_name}")
    for metric, value in metrics.items():
        print(f"  {metric}: {value:.4f}")

# -------------------------------
# Step 6: Plot Training History Comparison
# -------------------------------
plt.figure(figsize=(10,6))
for loss_name, history in histories.items():
    plt.plot(history['val_loss'], label=loss_name)
plt.title("Validation Loss Over Epochs for Different Loss Functions")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.legend()
plt.show()
```
